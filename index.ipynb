```python
import numpy as np
import pandas as pd

from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

import plotly.express as px
```

```python
def generate_data(n_samples=500, n_centers=4):
    b = np.linspace(start=0, stop=10, num=n_centers).reshape(-1, 1)

    centers = np.c_[b, b]
    X, _ = make_blobs(
        n_samples=n_samples,
        centers=centers,
        random_state=0,
        cluster_std=0.7)

    return X
```

```python
n_samples = 500
X = generate_data(n_samples)
```

```python
fig = px.scatter(x=X[:, 0], y=X[:, 1])
fig.update_layout(width=800)
fig.show()
```

```python
k_range = np.arange(1, 11)
labels = {}
for k in k_range:
    kmeans = KMeans(n_clusters=k)

    kmeans.fit(X)

    labels[f"label_k{k}"] = kmeans.labels_
```

```python
fig = px.scatter(x=X[:, 0], y=X[:, 1], color=labels['label_k4'])
fig.update_layout(width=800)
fig.show()
```

```python
df = pd.DataFrame(columns=['x1', 'x2'])
```

```python
df = pd.concat([df]*len(k_range))
```

```python
df[['k', 'label']] = pd.DataFrame(np.zeros(df.shape))
```

```python
df.head()
```

```python
i = 0

for k in k_range:
    df.iloc[i:i+n_samples, 2] = k * np.ones(n_samples)
    df.iloc[i:i+n_samples, 3] = labels[f"label_k{k}"]
```

```python
df.head()
```

```python
fig = px.scatter(
    data_frame=df,
    x='x1',
    y='x2',
    color='label',
    title='Agrupamento realizado pelo KMeans para cada k')

# fig.layout.updatemenus[0].buttons[0].args[1]['frame']['duration'] = 2000
fig.layout.coloraxis.showscale = False
fig.update_layout(width=800)
fig.show()
```
